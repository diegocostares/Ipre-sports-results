```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv("football_manager_2024_3_raw.csv",sep=";")
df1=df.drop(columns=['Unnamed: 75','Unnamed: 76','Unnamed: 77'])
df1 = df1.replace('-', np.nan)
df1.head()
```

```python
df2=df1
df2['Length'] = df2['Length'].str.extract('(\d+)').astype(int)
df2['Weight'] = df2['Weight'].str.extract('(\d+)').astype(int)
df2 = df2.rename(columns={'Length': 'Length (cms)','Weight': 'Weight (kg)'})

df2[['Caps', 'Goals']] = df2['Caps / Goals'].str.split(' / ', expand=True)
df2=df2.drop(columns=['Caps / Goals'])
caps = df2.pop("Caps")
goals = df2.pop("Goals")
df2.insert(6, "Goals", goals)
df2.insert(6, "Caps", caps)

df2['Sell value'] = df2['Sell value'].str.replace('€ ', '')
df2['Wages'] = df2['Wages'].str.replace('€ ', '')
df2['Wages'] = df2['Wages'].str.replace(' pw', '')

df2['Rel. clause'] = df2['Rel. clause'].str.replace('€ ', '')

df2.head()
```

```python
#Si el madrid presta a un jugador queda como que es del madrid pero en prestamo
#en el otro equipo, asi que borrare esa columna y cambiare el equipo por el prestamo
df3=df2.copy()
print(df3.iloc[10][["Name","Team","On loan at"]],"\n")

df3.loc[df3['On loan at'].notna(), 'Team'] = df3['On loan at']
print(df3.iloc[10][["Name","Team","On loan at"]])

df3=df3.drop(columns=["On loan at"])
df3[["Name","Team"]].head(15)
df3.head()
```

```python
#Al scrapear se me olvido agregarles la liga (la lista tiene de mas pero xd)
equipos_ligue_1 = ['PSG','Marseille','Lyon', 'AS Monaco','Rennes','Lille','OGC Nice','RC Lens','FC Nantes','Reims','Montpellier','Strasbourg','Brest','Toulouse','Lorient','Clermont','FC Metz','Havre AC']
bundesliga = ['Union Berlin','FC Bayern', 'RB Leipzig', 'Borussia Dortmund', 'Darmstadt 98','Bayer Leverkusen', 'VfL Wolfsburg', 'Eintracht Frankfurt', 'Borussia Mönchengladbach', 'FC Union Berlin', 'SC Freiburg', 'VfB Stuttgart', 'TSG Hoffenheim', 'Hertha BSC', 'FC Köln', 'FC Augsburg', 'FSV Mainz', 'Arminia Bielefeld', 'Werder Bremen', 'FC Schalke 04','Gladbach','VfL Bochum','FC Heidenheim']
premier = ['Manchester City', 'Manchester United', 'Liverpool', 'Chelsea', 'Arsenal', 'Tottenham Hotspur', 'Brentford','Nottingham Forest','Bournemouth','Luton Town','Leicester City', 'West Ham United', 'Aston Villa', 'Everton', 'Leeds United', 'Wolverhampton', 'Newcastle United', 'Crystal Palace', 'Southampton', 'Brighton & Hove Albion', 'Burnley', 'Fulham', 'West Bromwich Albion', 'Sheffield United']
laliga = ['Almería',"Cádiz",'Granada','Celta de Vigo','Rayo Vallecano','Mallorca','Alavés','Sevilla','Las Palmas','Osasuna','Getafe','Real Madrid','FC Barcelona','Girona','Atlético Madrid','Athletic Bilbao','Real Sociedad','Valencia','Real Betis','Villarreal']
SerieA = ['Juventus','Internazionale','AC Milan','Napoli','AS Roma','Lazio','Fiorentina','Atalanta','Torino','Sassuolo','Bologna','Udinese','Hellas Verona','Genoa','Monza','Cagliari','Salernitana','Empoli','Lecce','Frosinone']
def asignar_liga(equipo):
    if equipo in equipos_ligue_1:
        return 'Ligue 1'
    elif equipo in bundesliga:
        return 'Bundesliga'
    elif equipo in premier:
        return 'Premier League'
    elif equipo in laliga:
        return 'La Liga'
    elif equipo in SerieA:
        return 'Serie A'
    else:
        return np.nan

# Aplicar la función a la columna 'Team' y asignar los valores a la columna 'League'
df3['League'] = df3['Team'].apply(asignar_liga)

leagues = df3.pop("League")
df3.insert(10, "League", leagues)
df3.head()
```

```python
pd.set_option('display.max_rows', None)
df_nan = df3[df3['League'].isna()]
df_nan.head(15)
```

Estos son jugadores que estan en prestamo en otra ligas asi que los podemos eliminar

```python
df3 = df3.dropna(subset=['League'])
```

```python
null_counts_by_column = df3.isnull().sum(axis=0)
null_counts_by_column.reset_index().head(76)
```

Tenemos que las columnas Potential, On loan from, Loan ends y Rel. clause
tienen muchos NaN.
- El potential no es muy estimable y depende de un estudio profundo,ademas es una prediccion a futuro, por lo que podemos eliminarlo.
- On loan from, Loan ends tiene pocos datos pq pocos jugadores estan en prestamo, no afecta demasiado.
- Descubri que
solo la liga tiene Rel. clause, asi que se puede eliminar.
- Hay solo 24-28 jugadores que no tienen registro de sus ingresos, asi que los podemos eliminar porque deben ser de cantera.

```python
df4=df3.drop(columns=['Potential','Rel. clause'])
df4 = df4.dropna(subset=['Sell value',"Wages"])
df4[df4['Contract end'].isnull()].head(30)
null_counts_by_column = df4.isnull().sum(axis=0)
null_counts_by_column.reset_index().head(76)
```

La verdad son puros cadetes que de los equipos, asi que no tienen mucha relevancia para el estudio de grandes ligas.
